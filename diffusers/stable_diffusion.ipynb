```python
#step 1:Calculate the percentage of missing values in each column
missing_percentage=df.isnull().mean()* 100
missing_percentage
```

```python
#step 2: Decide whether to drop or impute missing values
x=30 #threshold percentage
for column, Percentage in missing_percentage.items():
    if Percentage >x:
        print(f"Column '{column}' has {Percenatge:.2f}% missing values and wil be dropped.")
        df.drop(column, axis=1, inplace=True)
    else:
        if df[column].dtype !='object':
            df[column].fillna(df[column].mean(), inplace=True)
print('Percentage of the missing values:',Percentage)
```

8. Find the number of videos published each month and arrange the months in a decreasing order based on the video count

```python
#CONVERT THE published_at COLUMN TO PANDAS DATETIME
df['published_at']=pd.to_datetime(df['published_at'])

#CREATE A NEW COLUMN AS published_month
df['published_month']=df['published_at'].dt.month
df['published_month']=df['published_month'].map(month_mapping)

#COUNT THE NUMBER OF VIDEOS PUBLISHED EACH MONTH
MONTH_COUNTS=df['published_month'].value_counts()

#SORT THE MONTH COUNTS IN DECREASING ORDER
SORTED_MONTH_COUNTS=MONTH_COUNTS.sort_values(ascending=False)

#PRINT THE SORTED MONTH COUNTS
SORTED_MONTH_COUNTS
```
